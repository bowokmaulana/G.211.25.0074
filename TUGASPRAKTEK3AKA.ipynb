pseudecode

```python
Algoritma Dijkstra(G, s):
Input  : Graf G(V, E), simpul awal s
Output : Jarak terpendek ke semua simpul

  1. Untuk setiap simpul v di V:
      dist[v] ← ∞
      prev[v] ← NIL

  2. dist[s] ← 0

  3. Buat priority queue Q
   Masukkan (0, s) ke Q

  4. Selama Q tidak kosong:
      (d, u) ← ambil simpul dengan jarak terkecil dari Q

      Jika d > dist[u]:
          lanjut

      Untuk setiap tetangga v dari u:
          jika dist[u] + w(u,v) < dist[v]:
              dist[v] ← dist[u] + w(u,v)
              prev[v] ← u
              masukkan (dist[v], v) ke Q

5. Kembalikan dist dan prev
```

coding

```python
import heapq

def dijkstra(graph, start):
    #inisialisasi jarak
    jarak = {node: float('inf') for node in graph}
    jarak[start] = 0

    pq = [(0, start)]  #(jarak, node)

    while pq:
        current_distance, current_node = heapq.heappop(pq)

        for neighbor, weight in graph[current_node]:
            distance = current_distance + weight

            if distance < jarak[neighbor]:
                jarak[neighbor] = distance
                heapq.heappush(pq, (distance, neighbor))

    return jarak


#contoh graph
graph = {
    'A': [('B', 3), ('C', 4)],
    'B': [('C', 6), ('D', 7)],
    'C': [('D', 1)],
    'D': []
}

hasil = dijkstra(graph, 'A')
print(hasil)
```
