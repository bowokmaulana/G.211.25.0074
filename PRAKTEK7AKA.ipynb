```python
# Fungsi PCA Sederhana Tanpa Library

def rata_rata(data):
    hasil = []
    jumlah_kolom = len(data[0])

    for j in range(jumlah_kolom):
        total = 0
        for i in range(len(data)):
            total += data[i][j]
        hasil.append(total / len(data))

    return hasil


def normalisasi(data, mean):
    data_baru = []

    for baris in data:
        sementara = []
        for i in range(len(baris)):
            sementara.append(baris[i] - mean[i])
        data_baru.append(sementara)

    return data_baru


def covariance_matrix(data):
    n = len(data)
    fitur = len(data[0])

    cov = []

    for i in range(fitur):
        baris = []
        for j in range(fitur):
            total = 0

            for k in range(n):
                total += data[k][i] * data[k][j]

            baris.append(total / (n - 1))

        cov.append(baris)

    return cov


def pca_manual(data):
    mean = rata_rata(data)

    data_normal = normalisasi(data, mean)

    cov = covariance_matrix(data_normal)

    print("Rata-rata Data:")
    print(mean)

    print("\nData Setelah Centering:")
    for x in data_normal:
        print(x)

    print("\nMatriks Kovarians:")
    for x in cov:
        print(x)

    return cov


# Contoh Data
dataset = [
    [2.5, 2.4],
    [0.5, 0.7],
    [2.2, 2.9],
    [1.9, 2.2],
    [3.1, 3.0],
    [2.3, 2.7],
    [2.0, 1.6],
    [1.0, 1.1],
    [1.5, 1.6],
    [1.1, 0.9]
]

pca_manual(dataset)
```
