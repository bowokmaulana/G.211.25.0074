```python
def create_node(key):
    return {
        "key": key,
        "left": None,
        "right": None
    }

def insert(root, key):
    if root is None:
        return create_node(key)

    if key < root["key"]:
        root["left"] = insert(root["left"], key)
    else:
        root["right"] = insert(root["right"], key)
    return root

def search(root, key):
    if root is None:
        return None

    if key == root["key"]:
        return root
    elif key < root["key"]:
        return search(root["left"], key)
    else:
        return search(root["right"], key)

root = None
data = [10, 20, 5, 15]

for x in data:
    root = insert(root, x)

cari = int(input("masukan angka yang di cari: "))

hasil = search(root, cari)

if hasil:
    print("Data ditemukan:", hasil["key"])
else:
    print("Data tidak ditemukan")
```

```python
def rb_search(node, key):
    while node is not None:
        if key == node.key:
            return node
        elif key < node.key:
            node = node.left
        else:
            node = node.right
    return None
```
