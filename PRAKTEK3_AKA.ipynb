euclidean distance

```python
import math

def euclidean_distance(a, b):
    jumlah = 0
    for i in range(len(a)):
        jumlah += (a[i] - b[i]) ** 2
    return math.sqrt(jumlah)

# contoh
x = [1, 2]
y = [4, 6]
print(euclidean_distance(x, y))   #output: 5.0
```

manhattan distance


```python
def manhattan_distance(a, b):
    total = 0
    for i in range(len(a)):
        total += abs(a[i] - b[i])
    return total

print(manhattan_distance([1, 2], [4, 6]))  #output: 7
```

cosine similarity

```python
import math

def cosine_similarity(a, b):
    dot = sum(a[i] * b[i] for i in range(len(a)))
    norm_a = math.sqrt(sum(x*x for x in a))
    norm_b = math.sqrt(sum(x*x for x in b))
    return dot / (norm_a * norm_b)

print(cosine_similarity([1, 2], [4, 6]))
```

hamming distance


```python
def hamming_distance(a, b):
    return sum(1 for i in range(len(a)) if a[i] != b[i])

print(hamming_distance("1011", "1001"))  #output: 1
```

permutasi


```python
def permutasi(arr, l, r):
    if l == r:
        print(arr)
    else:
        for i in range(l, r + 1):
            #swap
            arr[l], arr[i] = arr[i], arr[l]

            #rekursi
            permutasi(arr, l + 1, r)

            #backtrack (kembalikan seperti semula)
            arr[l], arr[i] = arr[i], arr[l]

#contoh penggunaan
data = [1, 2, 3]
permutasi(data, 0, len(data) - 1)
```

pseudecode

```python
Algoritma Dijkstra(G, s):
Input  : Graf G(V, E), simpul awal s
Output : Jarak terpendek ke semua simpul

  1. Untuk setiap simpul v di V:
      dist[v] ← ∞
      prev[v] ← NIL

  2. dist[s] ← 0

  3. Buat priority queue Q
   Masukkan (0, s) ke Q

  4. Selama Q tidak kosong:
      (d, u) ← ambil simpul dengan jarak terkecil dari Q

      Jika d > dist[u]:
          lanjut

      Untuk setiap tetangga v dari u:
          jika dist[u] + w(u,v) < dist[v]:
              dist[v] ← dist[u] + w(u,v)
              prev[v] ← u
              masukkan (dist[v], v) ke Q

5. Kembalikan dist dan prev
```

coding

```python
import heapq

def dijkstra(graph, start):
    #inisialisasi jarak
    jarak = {node: float('inf') for node in graph}
    jarak[start] = 0

    pq = [(0, start)]  #(jarak, node)

    while pq:
        current_distance, current_node = heapq.heappop(pq)

        for neighbor, weight in graph[current_node]:
            distance = current_distance + weight

            if distance < jarak[neighbor]:
                jarak[neighbor] = distance
                heapq.heappush(pq, (distance, neighbor))

    return jarak


#contoh graph
graph = {
    'A': [('B', 3), ('C', 4)],
    'B': [('C', 6), ('D', 7)],
    'C': [('D', 1)],
    'D': []
}

hasil = dijkstra(graph, 'A')
print(hasil)
```
